#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#importando base de dados

file_csv = 'b1abb33c4f.csv'

#carregando arquivo em um dataframe
data = pd.read_csv(file_csv)

# Exibir as primeiras linhas do DataFrame
print(data.head())
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Criar uma figura para os gráficos
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
axes = axes.flatten()  # Facilita o acesso aos eixos

# Variáveis para os gráficos
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
titulos = ['APGAR 1', 'APGAR 5', 'Sexo do Bebê', 'Gestação', 'Estado Civil da Mãe']

# Criar gráficos de barras para cada variável
for i, var in enumerate(variaveis):
    sns.countplot(data=data, x=var, ax=axes[i], palette="viridis")
    axes[i].set_title(titulos[i])
    axes[i].set_ylabel('Frequência')
    axes[i].set_xlabel(var)

# Remover o eixo vazio se houver
if len(variaveis) < len(axes):
    axes[-1].axis('off')

# Ajustar layout
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python


# Gráfico de barras
sns.barplot(data=data, x='CONSULTAS', y='APGAR5', ci=None, palette='viridis')

# Títulos e rótulos
plt.title('APGAR5 Médio por Número de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')

# Exibir o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Configuração do estilo
sns.set(style="whitegrid")

# Criar a figura e os eixos (2x2 painel)
fig, axes = plt.subplots(2, 2, figsize=(14, 10)) #Cria uma grade de 2x2 para os gráficos.
axes = axes.flatten()  # Facilita o acesso aos eixos

# Variáveis para os gráficos
variaveis = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']
titulos = ['Frequência por Sexo', 'Frequência por Gestação', 'Frequência por Estado Civil da Mãe', 'Frequência por Consultas']

# Criar gráficos de barras
for i, var in enumerate(variaveis):
    sns.countplot(data=data, x=var, ax=axes[i], palette='viridis')
    axes[i].set_title(titulos[i], fontsize=14)
    axes[i].set_xlabel(var, fontsize=12)
    axes[i].set_ylabel('Frequência', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)  # Rotacionar rótulos se necessário

# Ajustar layout
plt.tight_layout()
plt.show()
```
